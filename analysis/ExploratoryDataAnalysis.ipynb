# Word Cloud For Articles

```python
import ast 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator # for word cloud
```

```python
df = pd.read_csv("../data/fulltrain.csv", names=["Label", "Text"])
df["Text"] = df["Text"].apply(lambda x: x.lower())
df.head()
```

```python
# Create stopword list
stopwords = set(STOPWORDS)
stopwords.update(["said", "one", "now", "people", "will", 
                  "make", "time", "even", "may", "many",
                  "year", "day", "amp", "added", "say",
                  "much", "new", "according", "u", "think",
                  "way", "know", "old", "way", "going", "go",
                  "first", "right", "thing", "back", "still",
                  "can", "see", "come", "s", "percent",  "well",
                  "really", "every", "take",
                 ])
```

### All Articles

```python
words = ""

for x in df["Text"]:
    words += x
    
print ("There are {} words in the combination of all articles.".format(len(words)))
```

```python
flag_mask = np.array(Image.open("../imgs/newspaper.jpeg"))
mask_colors = ImageColorGenerator(flag_mask)

all_articles_wordcloud = WordCloud(stopwords=stopwords, mask=flag_mask, 
                             mode="RGBA", background_color="white", max_words=500, 
                             max_font_size=256, random_state=42,
                             width=flag_mask.shape[1],
                             height=flag_mask.shape[0],
                             color_func=mask_colors)
```

```python
all_articles_wordcloud.generate(words)

plt.imshow(all_articles_wordcloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout()

plt.savefig("../imgs/output/all_articles_wordcloud.png", format="png")
```
