```python
import numpy as np
import pandas as pd
from textblob import TextBlob
```

```python
training_data_filename = "../raw_data/fulltrain.csv"
df_train = pd.read_csv(training_data_filename, names=["label", "document"])
```

```python
# certain models cannot take negative values, so polarity [-1, 1] is padded to [0, 2]
df_train['polarity'] = df_train.document.apply(lambda x: TextBlob(x).sentiment.polarity + 1)
df_train['subjectivity'] = df_train.document.apply(lambda x: TextBlob(x).sentiment.subjectivity)
df_train
```

```python
X_train = pd.DataFrame(df_train, columns=['polarity', 'subjectivity'])
y_train = df_train["label"]
```
