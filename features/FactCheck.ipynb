```python
import time

X_train = []
y_train = []

correct = 0
wrong = 0

for index, row in df_train.iterrows():
    y_train.append(row["label"])
    
    text = row["document"]
    
    # Use OpenAI's GPT-3 to classify the text
    prompt = "Please classify the following text into one of the four classes: Satire, Hoax, Propaganda, or Reliable.\n\n" + text + "\n\nClass:"      
        
    completions = openai.Completion.create(
        engine="text-davinci-002",
        prompt=prompt,
        max_tokens=1,
        n=1,
        stop=None,
        temperature=0,
    )
    
    label = 0
    
    # Extract the classification from the response
    if len(completions.choices) > 0:
        classification = completions.choices[0].text.strip()
        print(classification)
        if (classification == 'Sat'):
            label = 1
        elif (classification == 'Ho'):
            label = 2
        elif (classification == 'Prop'):
            label = 3
        elif (classification == 'Rel'):
            label = 4
    
    X_train.append(label)
#     time.sleep(1)
    if (label == row["label"]):
        correct += 1
    else: 
        wrong += 1
        
print(correct)
print(wrong)
```
